```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

pd.set_option('display.max_columns', None)

from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv('./house-prices-advanced-regression-techniques/pre_processed.csv')
dataset.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(dataset.drop(['SalePrice'], axis=1), dataset.SalePrice, test_size=0.1, random_state=0)
X_train.head()
```

```python
feature_sel_model = SelectFromModel(Lasso(alpha=0.0005, random_state=0))
feature_sel_model.fit(X_train, y_train)
```

```python
feature_sel_model.get_support()
```

```python
selected_feature = X_train.columns[(feature_sel_model.get_support())]

print(f'Total features: {X_train.shape[1]}')
print(f'Selected features: {len(selected_feature)}')
print(f'Dropped features: {np.sum(feature_sel_model.get_support() == 0)}')
```

```python
selected_feature
```

```python
X_train = X_train[selected_feature]
X_test = X_test[selected_feature]
X_train.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

regression = LinearRegression()
regression.fit(X_train, y_train)
```

```python
predicted = regression.predict(X_test)
predicted
```
